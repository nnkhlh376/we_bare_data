# 03 - Baseline, LightGBM, Walk-forward Validation and Submission

Notebook này là pipeline mô hình chính cho bài toán dự báo `Revenue` và `COGS`.

## Mục tiêu

1. Xây dựng các baseline dự báo đơn giản như `lag-1` và `lag-365`.
2. Huấn luyện mô hình LightGBM cho hai target: `Revenue` và `COGS`.
3. Đánh giá mô hình bằng time-based split và walk-forward validation.
4. Kiểm soát leakage bằng cách chỉ sử dụng thông tin có sẵn trong quá khứ.
5. Tạo file `submission.csv` đúng định dạng yêu cầu.

## Thiết kế chống leakage

Final model chỉ sử dụng:
- Calendar features: tháng, ngày trong tuần, cuối tuần, ngày trong năm.
- Lag features của `Revenue` và `COGS`.
- Rolling statistics của `Revenue` và `COGS`.

Các biến vận hành như `sessions`, `orders`, `payments`, `returns`, `shipments` không được đưa trực tiếp vào final submission model vì trong giai đoạn test các giá trị cùng ngày này chưa được biết trước.

Với giai đoạn test, mô hình dự báo theo cơ chế recursive forecast: dự báo ngày hiện tại, sau đó dùng giá trị dự báo đó để tạo lag/rolling features cho các ngày tiếp theo. Seasonal baseline chỉ được xây dựng từ dữ liệu train đến ngày `2022-12-31`, đảm bảo không sử dụng thông tin tương lai.

```python
# 0. Setup, imports, paths
import os
import random
import warnings
from pathlib import Path

warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit

# --- REPRODUCIBILITY ---
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# --- MODEL AVAILABILITY CHECK ---
try:
    import lightgbm as lgb
    from lightgbm import LGBMRegressor
    HAS_LGB = True
except Exception:
    HAS_LGB = False
    print("LightGBM is not available, fallback to HistGradientBoostingRegressor.")
    from sklearn.ensemble import HistGradientBoostingRegressor

# --- ROBUST PATH RESOLUTION ---
def get_project_root() -> Path:
    """
    Identifies the project root directory ('FINAL').
    If the execution context is inside 'notebooks', it moves up to the parent.
    """
    # Get current working directory (absolute path)
    curr = Path.cwd().resolve()
    
    # If the current folder is 'notebooks', the root is the parent folder
    if curr.name == "notebooks":
        return curr.parent
    
    # Fallback logic: Search upwards for a directory containing 'dataset'
    for parent in [curr] + list(curr.parents):
        if (parent / "dataset").exists() and (parent / "notebooks").exists():
            return parent
            
    return curr

# Initialize PROJECT_ROOT
PROJECT_ROOT = get_project_root()

# Define paths relative to the actual PROJECT_ROOT
RAW_DIR = PROJECT_ROOT / "dataset" / "raw"
INTERIM_DIR = PROJECT_ROOT / "dataset" / "interim" / "daily"
FEATURE_PATH = INTERIM_DIR / "model_df.parquet"
OUT_DIR = PROJECT_ROOT / "outputs"

# Create output directory at the root level if it doesn't exist
OUT_DIR.mkdir(parents=True, exist_ok=True)

# --- DATA LOADING ---
if not FEATURE_PATH.exists():
    raise FileNotFoundError(
        f"Missing feature table at: {FEATURE_PATH}\n"
        "Please ensure your notebook (02) has generated the parquet file "
        "in the correct directory: root/dataset/interim/daily/"
    )

# Load and prepare data
df = pd.read_parquet(FEATURE_PATH)
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").reset_index(drop=True)

# --- FINAL STATUS CHECK ---
print(f"{'='*30}")
print(f"PROJECT_ROOT: {PROJECT_ROOT}")
print(f"RAW_DIR:      {'OK' if RAW_DIR.exists() else 'NOT FOUND'}")
print(f"FEATURE_PATH: {'OK' if FEATURE_PATH.exists() else 'NOT FOUND'}")
print(f"OUT_DIR:      {OUT_DIR}")
print(f"HAS_LGB:      {HAS_LGB}")
print(f"Data Shape:   {df.shape}")
print(f"Date Range:   {df['Date'].min().date()} -> {df['Date'].max().date()}")
print(f"{'='*30}")
```

```python
import pandas as pd
from pathlib import Path

# --- PHẦN SỬA: Logic tìm Root chuẩn xác hơn ---
def get_project_root() -> Path:
    """
    Ưu tiên nhảy ra khỏi thư mục 'notebooks' hoặc 'model' 
    để tìm thư mục gốc chứa 'dataset'.
    """
    curr = Path.cwd().resolve()
    
    # Nếu đang đứng trong thư mục notebooks hoặc model, nhảy ra ngoài 1 cấp trước
    if curr.name in ["notebooks", "model"]:
        return curr.parent
        
    # Nếu không, tìm thư mục cha nào chứa 'dataset'
    for _ in range(3):
        if (curr / "dataset").exists():
            return curr
        curr = curr.parent
    return Path.cwd().resolve().parent

PROJECT_ROOT = get_project_root()
RAW_DIR = PROJECT_ROOT / "dataset" / "raw"

# Kiểm tra file tồn tại
file_path = RAW_DIR / "sales.csv"
if not file_path.exists():
    raise FileNotFoundError(
        f"Vẫn không tìm thấy file tại: {file_path}\n"
        f"PROJECT_ROOT hiện tại đang là: {PROJECT_ROOT}\n"
        "Vui lòng kiểm tra xem bạn có đang chạy notebook trong thư mục 'notebooks' không."
    )

# 1. Load daily target data
sales = pd.read_csv(file_path)

# Ép kiểu Date
sales["Date"] = pd.to_datetime(sales["Date"], errors='coerce')

# Xử lý dữ liệu
sales = sales.dropna(subset=["Date"])
sales = sales.sort_values("Date").drop_duplicates("Date").reset_index(drop=True)

# Giữ lại giai đoạn training (<= 2022-12-31)
sales = sales.loc[sales["Date"] <= pd.Timestamp("2022-12-31")].copy()

print(f"--- PATH CHECK ---")
print(f"Project Root: {PROJECT_ROOT}")
print(f"Raw Dir:      {RAW_DIR}")
print(f"File Path:    {file_path}")
print(f"------------------")
print(f"Kích thước dữ liệu: {sales.shape}")
display(sales.head())
```

```python
# 2. Leakage-safe feature engineering
# All target lag/rolling features are shifted, so row t only uses information strictly before t.

def add_calendar_features(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out["year"] = out["Date"].dt.year
    out["month"] = out["Date"].dt.month
    out["quarter"] = out["Date"].dt.quarter
    out["dayofweek"] = out["Date"].dt.dayofweek
    out["dayofmonth"] = out["Date"].dt.day
    out["dayofyear"] = out["Date"].dt.dayofyear
    out["weekofyear"] = out["Date"].dt.isocalendar().week.astype(int)
    out["is_weekend"] = out["dayofweek"].isin([5, 6]).astype(int)
    
    # Cyclical encodings
    out["month_sin"] = np.sin(2 * np.pi * out["month"] / 12)
    out["month_cos"] = np.cos(2 * np.pi * out["month"] / 12)
    out["dow_sin"] = np.sin(2 * np.pi * out["dayofweek"] / 7)
    out["dow_cos"] = np.cos(2 * np.pi * out["dayofweek"] / 7)
    out["doy_sin"] = np.sin(2 * np.pi * out["dayofyear"] / 365.25)
    out["doy_cos"] = np.cos(2 * np.pi * out["dayofyear"] / 365.25)
    return out

LAGS = [1, 7, 14, 30, 60, 90, 180, 365]
ROLL_WINDOWS = [7, 14, 30, 60, 90]

def add_target_lag_features(df: pd.DataFrame) -> pd.DataFrame:
    out = df.sort_values("Date").copy()
    for target in ["Revenue", "COGS"]:
        shifted = out[target].shift(1)
        for lag in LAGS:
            out[f"{target.lower()}_lag_{lag}"] = out[target].shift(lag)
        for w in ROLL_WINDOWS:
            out[f"{target.lower()}_roll_mean_{w}"] = shifted.rolling(w, min_periods=max(2, w // 3)).mean()
            out[f"{target.lower()}_roll_std_{w}"] = shifted.rolling(w, min_periods=max(2, w // 3)).std()
            out[f"{target.lower()}_roll_min_{w}"] = shifted.rolling(w, min_periods=max(2, w // 3)).min()
            out[f"{target.lower()}_roll_max_{w}"] = shifted.rolling(w, min_periods=max(2, w // 3)).max()
    return out

# Build model table
model_df = add_calendar_features(sales)
model_df = add_target_lag_features(model_df)

# Drop rows without enough lag history. Keep from first available 365-day lag.
model_df = model_df.dropna().reset_index(drop=True)

feature_cols = [c for c in model_df.columns if c not in ["Date", "Revenue", "COGS"]]
feature_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(model_df[c])]

print("model_df:", model_df.shape)
print("Number of features:", len(feature_cols))
print(feature_cols[:30])
```

```python
# 3. Metrics and time split

def eval_reg(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    return {
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": rmse,
        "R2": r2_score(y_true, y_pred),
    }

train = model_df[model_df["Date"] <= "2020-12-31"].copy()
valid = model_df[
    (model_df["Date"] >= "2021-01-01") &
    (model_df["Date"] <= "2021-12-31")
].copy()
test_like = model_df[
    (model_df["Date"] >= "2022-01-01") &
    (model_df["Date"] <= "2022-12-31")
].copy()

print("train:", train["Date"].min(), train["Date"].max(), train.shape)
print("valid:", valid["Date"].min(), valid["Date"].max(), valid.shape)
print("test_like:", test_like["Date"].min(), test_like["Date"].max(), test_like.shape)
```

```python
# 4. Baselines: lag-1 and lag-365
baseline_rows = []

for target in ["Revenue", "COGS"]:
    prefix = target.lower()
    for split_name, split_df in [("valid", valid), ("test_like", test_like)]:
        for lag in [1, 365]:
            pred = split_df[f"{prefix}_lag_{lag}"].values
            baseline_rows.append({
                "target": target,
                "split": split_name,
                "model": f"lag_{lag}",
                **eval_reg(split_df[target], pred)
            })

baseline_result = pd.DataFrame(baseline_rows)
baseline_result.to_csv(OUT_DIR / "baseline_metrics.csv", index=False)
baseline_result
```

```python
# 5. Train LightGBM / fallback model with log1p target

def make_model():
    if HAS_LGB:
        return LGBMRegressor(
            objective="regression",
            n_estimators=3000,
            learning_rate=0.02,
            num_leaves=31,
            max_depth=-1,
            subsample=0.9,
            colsample_bytree=0.9,
            reg_alpha=0.05,
            reg_lambda=0.10,
            random_state=SEED,
            bagging_seed=SEED,
            feature_fraction_seed=SEED,
            data_random_seed=SEED,
            n_jobs=-1,
        )
    return HistGradientBoostingRegressor(
        loss="squared_error",
        learning_rate=0.04,
        max_iter=800,
        max_leaf_nodes=31,
        l2_regularization=0.05,
        random_state=SEED,
    )

models = {}
model_rows = []
valid_predictions = []

def fit_predict_model(target, X_train, y_train, X_eval=None, y_eval=None, X_pred=None):
    model = make_model()
    y_train_log = np.log1p(y_train)
    
    if HAS_LGB and X_eval is not None and y_eval is not None:
        model.fit(
            X_train, y_train_log,
            eval_set=[(X_eval, np.log1p(y_eval))],
            eval_metric="l1",
            callbacks=[lgb.early_stopping(200, verbose=False), lgb.log_evaluation(0)]
        )
    else:
        model.fit(X_train, y_train_log)
    
    if X_pred is None:
        return model, None
    pred = np.expm1(model.predict(X_pred))
    pred = np.clip(pred, 0, None)
    return model, pred

X_train = train[feature_cols]
X_valid = valid[feature_cols]
X_test_like = test_like[feature_cols]

for target in ["Revenue", "COGS"]:
    print(f"Training {target}")
    model, pred_valid = fit_predict_model(
        target,
        X_train, train[target],
        X_eval=X_valid, y_eval=valid[target],
        X_pred=X_valid
    )
    pred_test_like = np.expm1(model.predict(X_test_like))
    pred_test_like = np.clip(pred_test_like, 0, None)
    
    models[target] = model
    model_rows.append({"target": target, "split": "valid", "model": "lightgbm_log1p", **eval_reg(valid[target], pred_valid)})
    model_rows.append({"target": target, "split": "test_like", "model": "lightgbm_log1p", **eval_reg(test_like[target], pred_test_like)})
    
    tmp = valid[["Date", target]].copy()
    tmp["target"] = target
    tmp["prediction"] = pred_valid
    valid_predictions.append(tmp)

model_result = pd.DataFrame(model_rows)
model_result.to_csv(OUT_DIR / "model_holdout_metrics.csv", index=False)
pd.concat(valid_predictions).to_csv(OUT_DIR / "valid_predictions.csv", index=False)
model_result
```

```python
# 6. Walk-forward validation with early stopping from train fold only

def run_walk_forward(df_input, feature_cols, target_col, n_splits=4, horizon_days=365):
    df_wf = df_input.sort_values("Date").reset_index(drop=True).copy()
    X_all = df_wf[feature_cols]
    y_all = df_wf[target_col]
    d_all = df_wf["Date"]
    
    # Use fixed test_size to simulate annual forecasting windows
    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=horizon_days)
    rows = []
    preds = []
    
    for fold_id, (train_idx, test_idx) in enumerate(tscv.split(X_all), start=1):
        X_tr_full, X_te = X_all.iloc[train_idx], X_all.iloc[test_idx]
        y_tr_full, y_te = y_all.iloc[train_idx], y_all.iloc[test_idx]
        d_te = d_all.iloc[test_idx]
        
        # Baselines
        prefix = target_col.lower()
        for lag in [1, 365]:
            rows.append({
                "target": target_col,
                "fold": fold_id,
                "split_start": d_te.min(),
                "split_end": d_te.max(),
                "model": f"lag_{lag}",
                **eval_reg(y_te, df_wf.iloc[test_idx][f"{prefix}_lag_{lag}"])
            })
        
        # Split early-stopping validation from the tail of train fold only
        val_size = max(60, int(len(X_tr_full) * 0.15))
        X_fit, X_es = X_tr_full.iloc[:-val_size], X_tr_full.iloc[-val_size:]
        y_fit, y_es = y_tr_full.iloc[:-val_size], y_tr_full.iloc[-val_size:]
        
        wf_model, _ = fit_predict_model(target_col, X_fit, y_fit, X_eval=X_es, y_eval=y_es)
        pred_te = np.expm1(wf_model.predict(X_te))
        pred_te = np.clip(pred_te, 0, None)
        
        rows.append({
            "target": target_col,
            "fold": fold_id,
            "split_start": d_te.min(),
            "split_end": d_te.max(),
            "model": "lightgbm_log1p_ts_cv",
            **eval_reg(y_te, pred_te)
        })
        
        fold_pred = pd.DataFrame({
            "Date": d_te.values,
            "target": target_col,
            "fold": fold_id,
            "actual": y_te.values,
            "prediction": pred_te,
        })
        preds.append(fold_pred)
    
    return pd.DataFrame(rows), pd.concat(preds, ignore_index=True)

wf_rev, wf_pred_rev = run_walk_forward(model_df, feature_cols, "Revenue", n_splits=4, horizon_days=365)
wf_cogs, wf_pred_cogs = run_walk_forward(model_df, feature_cols, "COGS", n_splits=4, horizon_days=365)

walk_forward_metrics = pd.concat([wf_rev, wf_cogs], ignore_index=True)
walk_forward_predictions = pd.concat([wf_pred_rev, wf_pred_cogs], ignore_index=True)

walk_forward_metrics.to_csv(OUT_DIR / "walk_forward_metrics.csv", index=False)
walk_forward_predictions.to_csv(OUT_DIR / "walk_forward_predictions.csv", index=False)
walk_forward_metrics
```

```python
# 7. Train final models on all official training data through 2022-12-31
final_train = model_df[model_df["Date"] <= "2022-12-31"].copy()
X_final = final_train[feature_cols]

final_models = {}
for target in ["Revenue", "COGS"]:
    print(f"Training final {target} model")
    model, _ = fit_predict_model(target, X_final, final_train[target])
    final_models[target] = model

# Feature importance output, if LightGBM is available
if HAS_LGB:
    fi_rows = []
    for target, model in final_models.items():
        fi = pd.DataFrame({
            "feature": feature_cols,
            "importance": model.feature_importances_,
            "target": target,
        }).sort_values("importance", ascending=False)
        fi_rows.append(fi)
    feature_importance = pd.concat(fi_rows, ignore_index=True)
    feature_importance.to_csv(OUT_DIR / "feature_importance.csv", index=False)
    display(feature_importance.groupby("target").head(20))
```

```python
# 8. Train-only seasonal baseline for optional blending
# This profile uses only train history up to 2022-12-31, so it is leakage-safe.
train_sales = sales[sales["Date"] <= "2022-12-31"][["Date", "Revenue", "COGS"]].dropna().copy()
train_sales["year"] = train_sales["Date"].dt.year
train_sales["month"] = train_sales["Date"].dt.month
train_sales["day"] = train_sales["Date"].dt.day

annual = train_sales.groupby("year")[["Revenue", "COGS"]].mean()
full_years = annual.loc[2013:2022]
base_year = 2022
base_rev = annual.loc[base_year, "Revenue"]
base_cogs = annual.loc[base_year, "COGS"]

growth_rev = (full_years["Revenue"].iloc[-1] / full_years["Revenue"].iloc[0]) ** (1 / (len(full_years) - 1))
growth_cogs = (full_years["COGS"].iloc[-1] / full_years["COGS"].iloc[0]) ** (1 / (len(full_years) - 1))

annual_means = train_sales.groupby("year")[["Revenue", "COGS"]].transform("mean")
train_sales["rev_norm"] = train_sales["Revenue"] / annual_means["Revenue"]
train_sales["cogs_norm"] = train_sales["COGS"] / annual_means["COGS"]

seasonal_profile = train_sales.groupby(["month", "day"])[["rev_norm", "cogs_norm"]].mean().reset_index()
monthly_profile = train_sales.groupby("month")[["rev_norm", "cogs_norm"]].mean().reset_index()


def seasonal_baseline_for_date(dt):
    dt = pd.Timestamp(dt)
    m, d = dt.month, dt.day
    match = seasonal_profile[(seasonal_profile["month"] == m) & (seasonal_profile["day"] == d)]
    if len(match) == 0:
        match = monthly_profile[monthly_profile["month"] == m].rename(columns={"rev_norm": "rev_norm", "cogs_norm": "cogs_norm"})
    rev_norm = float(match["rev_norm"].iloc[0])
    cogs_norm = float(match["cogs_norm"].iloc[0])
    years_ahead = dt.year - base_year
    rev = base_rev * (growth_rev ** years_ahead) * rev_norm
    cogs = base_cogs * (growth_cogs ** years_ahead) * cogs_norm
    return max(rev, 0), max(cogs, 0)

print("growth_rev", growth_rev, "growth_cogs", growth_cogs)
```

```python
# 9. Recursive ML submission with seasonal blending
# Final file saved to: PROJECT_ROOT / outputs / submission.csv

import pandas as pd
import numpy as np
from pathlib import Path

# --- Path Management ---
# Using the robust detection logic to ensure we are at the /final/ level
def get_project_root() -> Path:
    curr = Path.cwd().resolve()
    for _ in range(4):
        if (curr / "dataset").exists():
            return curr
        curr = curr.parent
    return Path.cwd().resolve().parent

PROJECT_ROOT = get_project_root()
RAW_DIR = PROJECT_ROOT / "dataset" / "raw"
OUT_DIR = PROJECT_ROOT / "outputs"  # This points to D:\...\final\outputs
SAMPLE_PATH = RAW_DIR / "sample_submission.csv"
OUT_SUB = OUT_DIR / "submission.csv"

# Create directory if it doesn't exist
OUT_DIR.mkdir(parents=True, exist_ok=True)

# --- Data Preparation ---
sample_template = pd.read_csv(SAMPLE_PATH)
sample_template["Date"] = pd.to_datetime(sample_template["Date"])
sample_order = sample_template[["Date"]].copy()
test_dates = sample_order["Date"].tolist()

# Ensure 'sales' dataframe is available from previous steps
history = sales[["Date", "Revenue", "COGS"]].copy().sort_values("Date").reset_index(drop=True)

def make_recursive_feature_row(current_date, history):
    base = pd.DataFrame({
        "Date": [pd.Timestamp(current_date)],
        "Revenue": [np.nan],
        "COGS": [np.nan]
    })
    # Assumes add_calendar_features and LAGS/ROLL_WINDOWS are defined globally
    base = add_calendar_features(base)

    row = base.drop(columns=["Revenue", "COGS"]).iloc[0].to_dict()
    hist = history.sort_values("Date").reset_index(drop=True)

    for target in ["Revenue", "COGS"]:
        prefix = target.lower()
        values = hist[target]

        for lag in LAGS:
            row[f"{prefix}_lag_{lag}"] = values.iloc[-lag] if len(values) >= lag else values.iloc[-1]

        for w in ROLL_WINDOWS:
            window_vals = values.iloc[-w:] if len(values) >= w else values
            row[f"{prefix}_roll_mean_{w}"] = window_vals.mean()
            row[f"{prefix}_roll_std_{w}"] = window_vals.std() if len(window_vals) > 1 else 0
            row[f"{prefix}_roll_min_{w}"] = window_vals.min()
            row[f"{prefix}_roll_max_{w}"] = window_vals.max()

    row_df = pd.DataFrame([row])

    # Align with the feature columns used during training
    for col in feature_cols:
        if col not in row_df.columns:
            row_df[col] = 0

    row_df = row_df[feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
    return row_df

# --- Blending and Recursive Forecasting ---
ML_WEIGHT = 0.70  # 70% LightGBM + 30% seasonal baseline
blend_rows = []

for dt in test_dates:
    # 1. Generate features using updated history
    X_row = make_recursive_feature_row(dt, history)

    # 2. ML Prediction (Inverting log transformation)
    pred_rev_ml = np.expm1(final_models["Revenue"].predict(X_row))[0]
    pred_cogs_ml = np.expm1(final_models["COGS"].predict(X_row))[0]

    pred_rev_ml = max(float(pred_rev_ml), 0)
    pred_cogs_ml = max(float(pred_cogs_ml), 0)

    # 3. Seasonal Baseline Prediction
    pred_rev_seasonal, pred_cogs_seasonal = seasonal_baseline_for_date(dt)

    # 4. Blending
    pred_rev = ML_WEIGHT * pred_rev_ml + (1 - ML_WEIGHT) * pred_rev_seasonal
    pred_cogs = ML_WEIGHT * pred_cogs_ml + (1 - ML_WEIGHT) * pred_cogs_seasonal

    pred_rev = max(float(pred_rev), 0)
    pred_cogs = max(float(pred_cogs), 0)

    blend_rows.append({
        "Date": dt,
        "Revenue": pred_rev,
        "COGS": pred_cogs
    })

    # 5. Update history with the final blended prediction for the next recursive step
    history = pd.concat([
        history,
        pd.DataFrame([{
            "Date": dt,
            "Revenue": pred_rev,
            "COGS": pred_cogs
        }])
    ], ignore_index=True)

# --- Final Submission Formatting ---
submission = pd.DataFrame(blend_rows)
submission = sample_order.merge(submission, on="Date", how="left")

# Validations
assert submission["Revenue"].isna().sum() == 0
assert submission["COGS"].isna().sum() == 0
assert len(submission) == len(sample_template)
assert list(submission.columns) == ["Date", "Revenue", "COGS"]

# Save to CSV
submission["Date"] = submission["Date"].dt.strftime("%Y-%m-%d")
submission.to_csv(OUT_SUB, index=False)

print(f"Success! Final submission saved to: {OUT_SUB}")
print(f"ML_WEIGHT: {ML_WEIGHT}")
print(submission.head())
```
