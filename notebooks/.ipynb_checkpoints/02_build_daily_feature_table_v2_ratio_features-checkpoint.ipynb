# 02 - Xây dựng bảng đặc trưng cuối cùng theo ngày (Daily Feature Table)

## Mục tiêu
Hợp nhất các bảng daily/monthly thành một bảng feature duy nhất theo ngày để phục vụ EDA và train mô hình dự báo `Revenue`.

## Điểm đã chỉnh trong bản này
1. Join các bảng daily bằng `Date`.
2. Join `inventory_monthly_summary` theo nguyên tắc an toàn: **inventory tháng T được dùng cho tháng T+1** để tránh leakage.
3. Xử lý missing value theo bản chất nghiệp vụ:
   - biến đếm/tổng/sự kiện: fill 0
   - biến trung bình/tỷ lệ: fill median
4. Thêm các ratio features giống notebook model nâng cấp:
   - `conversion_rate`
   - `cancel_rate`
   - `return_rate`
   - `delivered_rate`
   - `items_per_order`
   - `discount_per_order`
   - `payment_per_order`
   - `pageviews_per_session`
5. Lưu output:
   - `dataset/interim/daily_feature_table.parquet`

## Lưu ý chống leakage
Không tạo các ratio dùng trực tiếp `Revenue` làm mẫu số/tử số, ví dụ `AOV = Revenue / total_orders`, trong file feature chính. Nếu cần AOV cho EDA thì tạo riêng ở notebook EDA, không đưa vào feature train model.


```python
import os
from pathlib import Path
import numpy as np
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "model" else Path.cwd().resolve()
IN_DIR = PROJECT_ROOT / "dataset" / "interim" / "daily"
OUT_PATH = PROJECT_ROOT / "dataset" / "interim" / "daily_feature_table.parquet"

def rd(name):
    p = IN_DIR / f"{name}.parquet"
    if not os.path.exists(p):
        raise FileNotFoundError(p)
    df = pd.read_parquet(p)
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    return df

sales_daily = rd("sales_daily")
web_traffic_daily = rd("web_traffic_daily")
orders_daily = rd("orders_daily")
payments_daily = rd("payments_daily")
order_items_daily = rd("order_items_daily")
promotions_daily = rd("promotions_daily")
returns_daily = rd("returns_daily")
reviews_daily = rd("reviews_daily")
shipments_daily = rd("shipments_daily")
inventory_monthly_summary = rd("inventory_monthly_summary")
```

```python
# ============================================================
# 2. Merge daily tables + create ratio and model-safe features
# ============================================================

# Merge all daily tables on Date
daily = sales_daily.copy()

for t in [
    web_traffic_daily,
    orders_daily,
    payments_daily,
    order_items_daily,
    promotions_daily,
    returns_daily,
    reviews_daily,
    shipments_daily
]:
    daily = daily.merge(t, on="Date", how="left")

# ============================================================
# Inventory: use previous month's inventory for current month
# Example: inventory 2020-06 -> used for all days in 2020-07
# ============================================================

inventory_monthly_summary = inventory_monthly_summary.copy()

inventory_monthly_summary["month_date"] = pd.to_datetime(
    inventory_monthly_summary["year"].astype(str)
    + "-"
    + inventory_monthly_summary["month"].astype(str)
    + "-01"
 )

inventory_monthly_summary["join_month"] = (
    inventory_monthly_summary["month_date"] + pd.DateOffset(months=1)
)

daily["join_month"] = daily["Date"].dt.to_period("M").dt.to_timestamp()

daily = daily.merge(
    inventory_monthly_summary.drop(columns=["year", "month", "month_date"]),
    on="join_month",
    how="left"
)

daily = daily.drop(columns=["join_month"])

# ============================================================
# Missing value policy
# ============================================================

if "sessions" in daily.columns:
    daily["has_web_traffic"] = daily["sessions"].notna().astype(int)
else:
    daily["has_web_traffic"] = 0

zero_fill_keywords = [
    "count", "total", "orders", "sessions", "visitors",
    "page_views", "quantity", "discount", "promo",
    "return", "refund", "shipment"
]

zero_fill_cols = [
    c for c in daily.columns
    if c != "Date" and any(k in c.lower() for k in zero_fill_keywords)
]

for c in zero_fill_cols:
    daily[c] = daily[c].fillna(0)

median_fill_cols = [
    c for c in daily.columns
    if c != "Date" and daily[c].dtype.kind in "biufc"
]

for c in median_fill_cols:
    daily[c] = daily[c].fillna(daily[c].median())

# ============================================================
# Ratio features
# ============================================================

def safe_div(numerator, denominator):
    if numerator not in daily.columns or denominator not in daily.columns:
        return pd.Series(0, index=daily.index)

    den = daily[denominator].replace(0, np.nan)
    out = daily[numerator] / den
    return out.replace([np.inf, -np.inf], np.nan).fillna(0)

daily["conversion_rate"] = safe_div("total_orders", "sessions")

daily["cancel_rate"] = safe_div("cancelled_orders", "total_orders")
daily["delivered_rate"] = safe_div("delivered_orders", "total_orders")
daily["return_rate"] = safe_div("returned_orders", "delivered_orders")

daily["items_per_order"] = safe_div("total_quantity_sold", "total_orders")
daily["discount_per_order"] = safe_div("total_discount_amount", "total_orders")
daily["payment_per_order"] = safe_div("total_payment_value", "total_orders")

daily["pageviews_per_session"] = safe_div("page_views", "sessions")

ratio_cols = [
    "conversion_rate",
    "cancel_rate",
    "delivered_rate",
    "return_rate",
    "items_per_order",
    "discount_per_order",
    "payment_per_order",
    "pageviews_per_session"
]

for c in ratio_cols:
    daily[c] = daily[c].replace([np.inf, -np.inf], np.nan).fillna(0)

# ============================================================
# Sort before lag/rolling
# ============================================================

daily = (
    daily
    .sort_values("Date")
    .drop_duplicates(subset=["Date"])
    .reset_index(drop=True)
)
# Remove duplicated columns if any
daily = daily.loc[:, ~daily.columns.duplicated()].copy()

print("Duplicated columns after cleanup:", daily.columns[daily.columns.duplicated()].tolist())
# ============================================================
# Model-safe lag/rolling features
# Only lag variables that would not be known before forecasting day D
# ============================================================

leak_prone_cols = [
    "sessions", "unique_visitors", "page_views", "bounce_rate",
    "avg_session_duration_sec",

    "total_orders", "cancelled_orders", "delivered_orders", "returned_orders",
    "total_payment_value", "avg_payment_value", "avg_installments",

    "total_quantity_sold", "total_discount_amount",
    "promo_line_count", "stacked_promo_count",

    "return_count", "total_refund_amount",
    "review_count", "avg_rating",

    "shipment_count", "avg_delivery_days", "avg_shipping_fee",

    "conversion_rate", "cancel_rate", "delivered_rate", "return_rate",
    "items_per_order", "discount_per_order", "payment_per_order",
    "pageviews_per_session"
 ]

# Add same-day order mix columns if they exist
leak_prone_cols += [
    c for c in daily.columns
    if c.startswith("orders_device_") or c.startswith("orders_source_")
]

# Remove duplicated names
leak_prone_cols = list(dict.fromkeys(leak_prone_cols))

new_features = {}

for c in leak_prone_cols:
    if c in daily.columns:
        s = daily[c]

        if isinstance(s, pd.DataFrame):
            s = s.iloc[:, 0]

        shifted = s.shift(1)
        new_features[f"{c}_lag_1"] = shifted
        new_features[f"{c}_rolling_7"] = shifted.rolling(7).mean()
        new_features[f"{c}_rolling_30"] = shifted.rolling(30).mean()

daily = pd.concat([daily, pd.DataFrame(new_features, index=daily.index)], axis=1)

# de-fragment dataframe
daily = daily.copy()

# ============================================================
# Final save
# ============================================================

daily.to_parquet(OUT_PATH, index=False)

print("Saved:", OUT_PATH)
print("Shape:", daily.shape)
print("Ratio features added:", ratio_cols)
print("Lag/rolling features added:", len(new_features))

daily.head()
```

```python
# Save a model-safe copy of the daily feature table for downstream training.
safe_drop_cols = [c for c in leak_prone_cols if c in daily.columns]
daily_model_safe = daily.drop(columns=safe_drop_cols, errors="ignore").copy()
daily_model_safe.to_parquet(OUT_PATH.with_name("daily_feature_table_model_safe.parquet"), index=False)

print("Saved model-safe table:", OUT_PATH.with_name("daily_feature_table_model_safe.parquet"))
print("Model-safe shape:", daily_model_safe.shape)
```

```python
# Data quality checks
assert daily["Date"].isna().sum() == 0, "Date has NaN"
assert daily["Date"].duplicated().sum() == 0, "Duplicate Date found"
assert "Revenue" in daily.columns, "Missing target Revenue"

ratio_cols = [
    "conversion_rate",
    "cancel_rate",
    "delivered_rate",
    "return_rate",
    "items_per_order",
    "discount_per_order",
    "payment_per_order",
    "pageviews_per_session"
]

missing_ratio_cols = [c for c in ratio_cols if c not in daily.columns]
assert len(missing_ratio_cols) == 0, f"Missing ratio cols: {missing_ratio_cols}"

numeric_bad = daily.select_dtypes(include=[np.number]).isin([np.inf, -np.inf]).sum().sum()
assert numeric_bad == 0, "There are inf values in numeric columns"

print("DQ checks passed")
print("Date range:", daily["Date"].min(), "->", daily["Date"].max())
print("Rows:", len(daily))
daily[["Date", "Revenue"] + ratio_cols].head(10)
```

```python
from pathlib import Path

CWD = Path.cwd()
PROJECT_ROOT = CWD.parent if CWD.name == "model" else CWD

OUT_PATH = PROJECT_ROOT / "dataset" / "interim" / "daily" / "model_df.parquet"
OUT_PATH.parent.mkdir(parents=True, exist_ok=True)

daily_model_safe.to_parquet(OUT_PATH, index=False)

print("Saved:", OUT_PATH)
print("Shape:", daily_model_safe.shape)
```
