```python
from pathlib import Path
import os

import pandas as pd

# ============================================================
# 0. Project paths
# ============================================================

def find_project_root(start: Path = Path.cwd()) -> Path:
    """
    Find project root by walking parents.
    A valid repo should contain dataset/raw or data/raw.
    """
    env_root = os.getenv("PROJECT_ROOT")
    if env_root:
        return Path(env_root).expanduser().resolve()

    start = start.resolve()
    candidates = [start] + list(start.parents)

    for p in candidates:
        if (p / "dataset" / "raw").exists():
            return p
        if (p / "data" / "raw").exists():
            return p
        if (p / ".git").exists():
            return p

    return start


PROJECT_ROOT = find_project_root()

RAW_DIR_CANDIDATES = [
    PROJECT_ROOT / "dataset" / "raw",
    PROJECT_ROOT / "data" / "raw",
    PROJECT_ROOT,
]

RAW_DIR = next((p for p in RAW_DIR_CANDIDATES if p.exists()), PROJECT_ROOT)
INTERIM_DIR = PROJECT_ROOT / "dataset" / "interim"
INTERIM_DIR.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_DIR:", RAW_DIR)
print("INTERIM_DIR:", INTERIM_DIR)
```

```python
# ============================================================
# 1. Clean raw CSV -> parquet in dataset/interim
# ============================================================

DATE_COLUMNS = {
    "sales": ["date"],
    "orders": ["order_date"],
    "shipments": ["ship_date", "delivery_date"],
    "returns": ["return_date"],
    "reviews": ["review_date"],
    "web_traffic": ["date"],
    "inventory": ["snapshot_date"],
    "promotions": ["start_date", "end_date"],
    "payments": ["payment_date", "approved_date"],
    "order_items": ["shipping_limit_date"],
}


def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns.str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
    )
    return df


def clean_table(df: pd.DataFrame, table_name: str) -> pd.DataFrame:
    df = normalize_columns(df)
    df = df.drop_duplicates().reset_index(drop=True)
    for col in DATE_COLUMNS.get(table_name, []):
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")
    return df


def load_csv(path: Path) -> pd.DataFrame:
    try:
        return pd.read_csv(path)
    except UnicodeDecodeError:
        return pd.read_csv(path, encoding="latin-1")


raw_csv_files = sorted(RAW_DIR.glob("*.csv"))
if not raw_csv_files:
    raise FileNotFoundError(f"No CSV files found in {RAW_DIR}")


written = []
for csv_path in raw_csv_files:
    table_name = csv_path.stem
    df = load_csv(csv_path)
    df = clean_table(df, table_name)
    out_path = INTERIM_DIR / f"{table_name}_base.parquet"
    df.to_parquet(out_path, index=False)
    written.append((table_name, len(df), out_path))


print("Wrote parquet files:")
for table_name, nrows, out_path in written:
    print(f"- {table_name:20s} | rows={nrows:8d} | {out_path}")
```
